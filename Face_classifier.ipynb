```python
import face_recognition
import cv2
import os
from utils import create_csv

# The output video
fourcc = cv2.VideoWriter_fourcc(*'XVID')
output_movie = cv2.VideoWriter( '/Users/DZ/Face Recognation/Output/money_heist_output_final.avi', fourcc, 30, (1280, 720))

# Open the input movie file
input_movie = cv2.VideoCapture("/Users/DZ/Face Recognation/test_video.mp4")
length = int(input_movie.get(cv2.CAP_PROP_FRAME_COUNT))

# Load some sample pictures and learn how to recognize them.
berlin_image = face_recognition.load_image_file("/Users/DZ/Face Recognation/Face/berlin.jpg")
berlin_face_encoding = face_recognition.face_encodings(berlin_image)[0]

nairobi_image = face_recognition.load_image_file("/Users/DZ/Face Recognation/Face/nairobi1.jpg")
nairobi_face_encoding = face_recognition.face_encodings(nairobi_image)[0]

professor_image = face_recognition.load_image_file("/Users/DZ/Face Recognation/Face/professor.jpg")
professor_face_encoding = face_recognition.face_encodings(professor_image)[0]

monica_image = face_recognition.load_image_file("/Users/DZ/Face Recognation/Face/monica.jpg")
monica_face_encoding = face_recognition.face_encodings(monica_image)[0]

denver_image = face_recognition.load_image_file("/Users/DZ/Face Recognation/Face/denver.jpg")
denver_face_encoding = face_recognition.face_encodings(denver_image)[0]

helsinki_image = face_recognition.load_image_file("/Users/DZ/Face Recognation/Face/helsinki.jpg")
helsinki_face_encoding = face_recognition.face_encodings(helsinki_image)[0]

rio_image = face_recognition.load_image_file("/Users/DZ/Face Recognation/Face/rio.jpg")
rio_face_encoding = face_recognition.face_encodings(rio_image)[0]

tokyo_image = face_recognition.load_image_file("/Users/DZ/Face Recognation/Face/tokyo.jpg")
tokyo_face_encoding = face_recognition.face_encodings(tokyo_image)[0]

raquel_image = face_recognition.load_image_file("/Users/DZ/Face Recognation/Face/raquel.jpg")
raquel_face_encoding = face_recognition.face_encodings(raquel_image)[0]

moscow_image = face_recognition.load_image_file("/Users/DZ/Face Recognation/Face/moscow.jpg")
moscow_face_encoding = face_recognition.face_encodings(moscow_image)[0]

known_faces = [
    berlin_face_encoding,
    nairobi_face_encoding,
    professor_face_encoding,
    monica_face_encoding,
    denver_face_encoding,
    helsinki_face_encoding,
    rio_face_encoding,
    tokyo_face_encoding,
    raquel_face_encoding,
    moscow_face_encoding
]

# Initialize some variables
face_locations = []
face_encodings = []
face_names = []
frame_number = 0

current_path = os.getcwd()

counter = 0
counter1 = 0

while True:
    # Grab a single frame of video
    ret, frame = input_movie.read()
    frame_number += 1

    # Quit when the input video file ends
    if not ret:
        break

    # Find all the faces and face encodings in the current frame of video
    face_locations = face_recognition.face_locations(frame)
    face_encodings = face_recognition.face_encodings(frame, face_locations)

    face_names = []
    for face_encoding in face_encodings:
        # See if the face is a match for the known face(s)
        match = face_recognition.compare_faces(
            known_faces, face_encoding, tolerance=0.6)

        # If you had more than 2 faces, you could make this logic a lot prettier
        # but I kept it simple for the demo
        name = None
        if match[0]:
            name = "Berlin"
        elif match[1]:
            name = "Nairobi"
        elif match[2]:
            name = "Professor"
        elif match[3]:
            name = "Monica"
        elif match[4]:
            name = "Denver"
        elif match[5]:
            name = "Helsinki"
        elif match[6]:
            name = "Rio"
        elif match[7]:
            name = "Tokyo"
        elif match[8]:
            name = "Raquel"
        elif match[9]:
            name = "Moscow"

        face_names.append(name)

    # Label the results
    for (top, right, bottom, left), name in zip(face_locations, face_names):
        if not name:
            continue

        # Draw a box around the face
        cv2.rectangle(frame, (left, top), (right, bottom), (0, 0, 255), 2)
        #cv2.circle(frame, (left, top), 2, (0, 255, 0), -1) 
        crop_img = frame[top:bottom, left:right]
        if(name == "Nairobi"):
                cv2.imwrite(os.path.join(os.path.expanduser('~'),'Face_data_nairobi',"nairobi"+str(counter)+".bmp"), crop_img)
                counter = counter + 1
            ##elif(name == "Nairobi"):
            #cv2.imwrite("/Users/sreenandk/Desktop/ml/Money/face_database/Nairobi/" + "nairobi"+str(counter1)+".png", crop_img)
            #counter1 = counter1 + 1

        # Draw a label with a name below the face
        cv2.rectangle(frame, (left, bottom - 35),(right, bottom), (0, 0, 255), cv2.FILLED)
        #cv2.circle(frame, (left, bottom), 2, (0, 255, 0), -1) 
        font = cv2.FONT_HERSHEY_DUPLEX
        cv2.putText(frame, name, (left + 6, bottom - 6), font, 1.0, (255, 255, 255), 2)

    # Write the resulting image to the output video file
    output_movie.write(frame)
    print("Writing frame {} / {}".format(frame_number, length))

    cv2.imshow('face_recog_crop', frame)
    # Hit 'q' on the keyboard to quit!
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# All done!
input_movie.release()
cv2.destroyAllWindows()
create_csv.CreateCsv("/Users/DZ/Face Recognation/Face/face_database/")
```
